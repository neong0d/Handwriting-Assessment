#### **CONTAINS SOME SAMPLE VISUALISATIONS OF THE HANDWRITTEN CHARACTERS**

```python
import numpy as np
import matplotlib.pyplot as plt
import pathlib, os
```

**SUPERIMPOSED PLOT OF ALL 300 SAMPLES OF TWO CHARACTERS: 'A' & 'MA'**

```python
directory = pathlib.Path().parent.resolve()
A_dir = os.path.join(directory, '90point/A/')
MA_dir = os.path.join(directory, '90point/MA')
```

```python
# superimposed A
fig = plt.figure(figsize=(8,8)) 
ax = fig.add_subplot()

for file in os.listdir(A_dir):
    data = np.loadtxt(os.path.join(A_dir, file))
    x,y,z = np.split(data, 3, axis=1)
    x,y = x.flatten(), y.flatten()
    y = y*-1+512

    ax.scatter(x, y, c='green', s=10)
    ax.set_xlim(-20,532)
    ax.set_ylim(-20,532)
plt.axis('off')
plt.show()
```

```python
# superimposed MA
fig = plt.figure(figsize=(8,8)) 
ax = fig.add_subplot()

for file in os.listdir(MA_dir):
    data = np.loadtxt(os.path.join(MA_dir, file))
    x,y,z = np.split(data, 3, axis=1)
    x,y = x.flatten(), y.flatten()
    y = y*-1+512

    ax.scatter(x, y, c='green', s=10)
    ax.set_xlim(-20,532)
    ax.set_ylim(-20,532)
plt.axis('off')
plt.show()
```
