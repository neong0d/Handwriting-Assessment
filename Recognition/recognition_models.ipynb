```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset.csv')
df
```

```python
df.describe()
```

### **PREPARING DATA FOR MODELS**  
- encoding text labels into numerics
- splitting training & testing set
- scaling feature vectors

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
labels_encoded = encoder.fit_transform(df['label'].values)
labels_encoded
```

```python
len(set(labels_encoded))
```

```python
df['label_en'] = labels_encoded
df_prepared = df.copy()
df_prepared = df_prepared.drop('label', axis=1)
df_prepared
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(
    n_splits=1, test_size=0.2, random_state=42
)
for train_index, test_index in split.split(df_prepared, df_prepared['label_en']):
    strat_train_set = df_prepared.loc[train_index]
    strat_test_set = df_prepared.loc[test_index]
```

```python
strat_train_set
```

```python
feat_train, label_train = strat_train_set.drop('label_en', axis=1), strat_train_set['label_en']
feat_test, label_test = strat_test_set.drop('label_en', axis=1), strat_test_set['label_en']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feat_train_scaled = scaler.fit_transform(feat_train.astype(np.float64))
feat_test_scaled = scaler.fit_transform(feat_test.astype(np.float64))
```

### **COMPARING 4 CLASSIFIERS ON TRAINING SET**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (
    precision_score, recall_score,
    f1_score, confusion_matrix, 
    accuracy_score, classification_report
)
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
comparison = pd.DataFrame(columns=['model', 'accuracy', 'precision', 'f1'])
models = [
    RandomForestClassifier(random_state=42),
    SVC(),
    #MultinomialNB(),
    KNeighborsClassifier(),
]
cv = 4
for model in models:
    score = cross_val_score(
        model, feat_train_scaled, label_train,
        cv=cv, scoring='accuracy'
    )
    predictions = cross_val_predict(
        model, feat_train_scaled, label_train, cv=cv
    )
    accuracy = sum(score)/4
    precision = precision_score(label_train, predictions, average='weighted')
    f1 = f1_score(label_train, predictions, average='weighted')
    comparison.loc[len(comparison.index)] = [
        model.__class__.__name__, accuracy, precision, f1
    ]

comparison
```

```python
model = MultinomialNB()
score = cross_val_score(
        model, feat_train, label_train,
        cv=cv, scoring='accuracy'
    )
predictions = cross_val_predict(
        model, feat_train, label_train, cv=cv
    )
accuracy = sum(score)/4
precision = precision_score(label_train, predictions, average='weighted')
f1 = f1_score(label_train, predictions, average='weighted')
comparison.loc[len(comparison.index)] = [
        model.__class__.__name__, accuracy, precision, f1
    ]
```

```python
comparison
```

### **FINAL MODEL -> SVC**
- testing on test set
- generating classification report, confusion matrix

```python
model = SVC()
model.fit(feat_train_scaled, label_train)
predictions = model.predict(feat_test_scaled)
```

```python
conf_mat = confusion_matrix(label_test, predictions)
fig, ax = plt.subplots(figsize=(10,10))
ax.matshow(conf_mat, cmap=plt.cm.gray)
ax.set_xticks(np.arange(50))
ax.set_yticks(np.arange(50))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
plt.show()
```

```python
row_sums = conf_mat.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mat/row_sums

np.fill_diagonal(norm_conf_mx, 0)
fig, ax = plt.subplots(figsize=(16,16))
ax.matshow(norm_conf_mx, cmap=plt.cm.gray)
ax.set_xticks(np.arange(50))
ax.set_yticks(np.arange(50))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
plt.show()
```

```python
from sklearn import metrics
print(metrics.classification_report(label_test, predictions, target_names=labels))
```

### **HYPERPARAMETER TUNING OF SVC**

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform
param_distributions = {"gamma": reciprocal(0.001, 0.1), "C": uniform(1, 10)}
```

```python
rscv = RandomizedSearchCV(SVC(gamma='scale'), param_distributions)
rscv.fit(feat_train_scaled, label_train)
```

```python
rscv.best_estimator_
```

```python
model = rscv.best_estimator_
model.fit(feat_train_scaled, label_train)
```

```python
tuned_pred = model.predict(feat_test_scaled)

accuracy = accuracy_score(label_test, tuned_pred)
precision = precision_score(label_test, tuned_pred, average='weighted')
f1 = f1_score(label_test, tuned_pred, average='weighted')

print(f'accuracy: {accuracy}\nprecision: {precision}\nf1: {f1}')
```

```python
conf_mat = confusion_matrix(label_test, tuned_pred)
fig, ax = plt.subplots(figsize=(12,12))
ax.matshow(conf_mat, cmap=plt.cm.gray)
ax.set_xticks(np.arange(50))
ax.set_yticks(np.arange(50))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
plt.show()
```

```python
row_sums = conf_mat.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mat/row_sums

np.fill_diagonal(norm_conf_mx, 0)
fig, ax = plt.subplots(figsize=(16,16))
ax.matshow(norm_conf_mx, cmap=plt.cm.gray)
ax.set_xticks(np.arange(50))
ax.set_yticks(np.arange(50))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
plt.show()
```
